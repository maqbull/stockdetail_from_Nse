# APPROACH :-

I have imported NSE from python library called nsetools which provides realtime data from nse website ,and then created empty list after this defined function which take argument as stock name then provides data required .

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
from nsetools import Nse
from pandas import DataFrame
nse = Nse()

DataRequired = []

def requiredData(code):

    quote = nse.get_quote(code)
    # dictionary which store data fetches from nse

    DesiredData= {}
    # empty dictionary

   # copied required data from quote
    DesiredData['Name'] = code
    DesiredData['Open'] = quote['basePrice']
    DesiredData['High'] = quote['dayHigh']
    DesiredData['Low'] = quote['dayLow']
    DesiredData['Close'] = quote['closePrice']
    DesiredData['Volume'] = quote['quantityTraded']
    DesiredData['Delivery Data'] = quote['deliveryQuantity']
    DesiredData['Delivery Percentage'] = quote['deliveryToTradedQuantity']
    # appended dictionary to the list
    DataRequired.append(DesiredData)


requiredData('reliance')
requiredData('hdfcbank')
requiredData('adaniports')
requiredData('itc')
requiredData('sbin')
requiredData('ioc')
requiredData('rblbank')
# passing the argument to function

#print(DataRequired) 


df = pd.DataFrame(DataRequired)
df.head(10)
```
